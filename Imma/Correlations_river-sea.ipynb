## Obtaining data from rescue actions <a class="anchor" id="1"></a>

```python
accions_salvament = pd.read_csv(direc+'Actuacions_en_salvaments_al_medi_natural_dels_Bombers_de_la_Generalitat_20231109.csv')
accions_filtrades= accions_salvament[(accions_salvament["ANY"]<2023) & (accions_salvament["ANY"]>2017)]   #AND & OR |

# show loaded table:
accions_filtrades['TAL_NOM_ALARMA'].unique()
```

```python
#MAR_RIU - Salvament i/o recerca i/o rescat en medi fluvial (rius,barrancs, llacs) - Salvament i/o recerca i/o rescat en medi marítim - Recerca i/o rescat en medi marítim - Recerca i/o rescat en medi fluvial (rius, barrancs, llacs) 
accions_descartades_riumar = ['Salvament i/o rescat de muntanya',
       'Recerca persones perdudes o desaparegudes',
       'Salvament i/o rescat en coves i pous', 'Rescat de muntanya',
       'Recerca persones perdudes', 'Rescat en coves i pous']

salvam_riu = accions_filtrades[~accions_filtrades['TAL_NOM_ALARMA'].isin(accions_descartades_riumar)]

salvam_riu= salvam_riu.reset_index(drop=True)

salvam_riu

resultado = salvam_riu.loc[salvam_riu['MUNICIPI'] == 'Llançà']
resultado
```

```python
#Turn "municipi" in capital letters

Salvam_riu=salvam_riu.copy()

from unidecode import unidecode
def quitar_acentos(cadena):
    # Reemplazar 'ç' por un marcador temporal
    cadena_con_marcador = cadena.replace('ç', '___cedilla___')
    # Aplicar unidecode y luego revertir el marcador temporal
    cadena_sin_acentos = unidecode(cadena_con_marcador).replace('___cedilla___', 'ç')
    
    return cadena_sin_acentos

Salvam_riu['MUNICIPI'] = Salvam_riu['MUNICIPI'].apply(quitar_acentos)


Salvam_riu['MUNICIPI'] = Salvam_riu['MUNICIPI'].str.upper()

Salvam_riu

# resultado = Salvam_riu.loc[Salvam_riu['MUNICIPI'] == 'LLANÇA']
# resultado
```

```python
#treure totes les columnes que no m'ajuden
#salvam_riu.pop('')
riu_clean = Salvam_riu.copy()
#Salvam_riu.columns.tolist()

riu_clean.pop('TAL_COD_ALARMA1') 
riu_clean.pop('TAL_COD_ALARMA2')
riu_clean.pop('TGA_NOM_GRUPO')
riu_clean.pop('CODI_INE')
riu_clean.pop('NOM_POBLACIO')
riu_clean.pop('ACT_NUM_ACTUACIO')
riu_clean.pop('NOM REGIO')

riu_clean=riu_clean[(riu_clean['VALIDAT'] == 'S') | (riu_clean['VALIDAT'] == 'SI') ]
salva_clean_riu = riu_clean.copy()
salva_clean_riu['ANY-MES'] = pd.to_datetime(salva_clean_riu['ANY'].astype(str) + '-' + salva_clean_riu['MES'].astype(str), format='%Y-%m')
salva_clean_riu.pop('ANY')
salva_clean_riu.pop('MES')
salva_clean_riu.pop('ACT_DAT_ACTUACIO')
salva_clean_riu.pop('VALIDAT')
salva_clean_riu.pop('CODI_COMARCA')
salva_clean_riu.pop('CODI_REGIO')
salvament = salva_clean_riu.copy()
salvament
```

### Grouped recue - sea-river-like  <a class="anchor" id="1.1"></a>

```python
#Grouping all recue actions in the same town in the same month and year
sal = salvament.copy()
sal=pd.DataFrame(sal.groupby(['ANY-MES','MUNICIPI','NOM_COMARCA','LONGITUD','LATITUD','GEOREFERÈNCIA','ACT_X_UTM','ACT_Y_UTM','TAL_NOM_ALARMA'])['ANY-MES'].count())
sal= sal.rename(columns={'ANY-MES':"SALVAMENTS"})
sal = sal.reset_index()
sal_grouped=sal.copy()

sal_grouped
# inc_riu_grouped = pd.DataFrame (inc_riu.groupby(['ANY-MES','PROVINCIA','MUNICIPI','COMARCA'])['INCIDENTS'].sum())
```

## Importing data from 112 incidents <a class="anchor" id="2"></a>

```python
trucades_salvament= pd.read_csv(direc + 'Dades_d_incidents_operatius_gestionats_pel_CAT112_20231110.csv' )
trucades_filtrades= trucades_salvament[(trucades_salvament["ANY"]<2023) & (trucades_salvament["ANY"]>2017)]

trucades_filtrades['TIPUS'].unique()

trucades_filtrades
```

```python
#Incidents utils comparables
incidents_descartats=['Seguretat','Trànsit','Civisme']

incidents = trucades_filtrades[~trucades_filtrades['TIPUS'].isin(incidents_descartats)]
incidents = incidents.reset_index(drop=True)
incidents
```

```python
#focusing in river sea actions:

incidents_desc_riumar = ['Fuita', 'Altres incidències', 'Incendi', 'Medi ambient']
         

incidents_riu = incidents[~incidents['TIPUS'].isin(incidents_desc_riumar)]
incidents_riu['TIPUS'].unique()
incidents_riu['ANY-MES'] = pd.to_datetime(incidents_riu['ANY'].astype(str) + '-' + incidents_riu['MES'].astype(str), format='%Y-%m')
incidents_riu.pop('ANY')
incidents_riu.pop('MES')
inc_riu = incidents_riu
inc_riu
```

### Obtaining data of river-sea-like rescues of the incidents at each town each month <a class="anchor" id="2.1"></a>

```python
#Grouping all incidenst in the same town in the same month and year

inc_riu_grouped = pd.DataFrame (inc_riu.groupby(['ANY-MES','PROVINCIA','MUNICIPI','COMARCA'])['INCIDENTS'].sum())

inc_riu_grouped= inc_riu_grouped.reset_index()

inc_riu_grouped.drop_duplicates()
```

## Merging data <a class="anchor" id="3"></a>

```python
#Merging rescue actions with incidents of river and sea nature

merge_sea = pd.merge(inc_riu_grouped,sal_grouped, on=['MUNICIPI','ANY-MES'])
merge_sea
```
