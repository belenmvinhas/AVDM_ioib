# Incident distribution per region

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
direc = '../dades/'
```

## Importing incidents and filtering

```python
incidents_salvament= pd.read_csv(direc + 'Dades_d_incidents_operatius_gestionats_pel_CAT112_20231110.csv' )
incidents_salvament= incidents_salvament[(incidents_salvament["ANY"]<=2022) & (incidents_salvament["ANY"]>=2018)]


incidents_salvament.head()
```

```python
grouped_df=incidents_salvament.copy()

grouped_df=pd.DataFrame(grouped_df.groupby(['COMARCA','TIPUS'])['INCIDENTS'].mean()).reset_index()

replacement_dict = {'Accident': 'Non-Traffic Accident', 'Altres incidències': 'Other incidents', 'Assistència sanitària': 'Medical Assistance','Civisme':'Civility','Fuita (aigua, gas, altres)':'Leak','Incendi':'Fire', 'Medi ambient':'Environment',
       'Meteorologia':'Meteorology', 'Seguretat':'Security', 'Trànsit':'Traffic'}

# Replace values in the 'Fruit' column using the dictionary
grouped_df['TIPUS'].replace(replacement_dict, inplace=True)

grouped_df['TIPUS'].unique()
```

## Creaiting one data frame per region

The new order is defined so this typologies will be together in the pie chart and the percentage it represents can be more easely seen

```python
new_order = [0,5,7,2,1,3,4,6,8,9]
noguera = grouped_df[grouped_df['COMARCA']== 'NOGUERA'].iloc[new_order].reset_index(drop=True)
osona = grouped_df[grouped_df['COMARCA']== 'OSONA'].iloc[new_order].reset_index(drop=True)
pallars_jussa = grouped_df[grouped_df['COMARCA']== 'PALLARS JUSSA'].iloc[new_order].reset_index(drop=True)
pallars_sobira = grouped_df[grouped_df['COMARCA']== 'PALLARS SOBIRA'].iloc[new_order].reset_index(drop=True)
pla_durgell = grouped_df[grouped_df['COMARCA']== "PLA D'URGELL"].iloc[new_order].reset_index(drop=True)

pla_de_lestany = grouped_df[grouped_df['COMARCA']== "PLA DE L'ESTANY"].iloc[new_order].reset_index(drop=True)
priorat = grouped_df[grouped_df['COMARCA']== 'PRIORAT'].iloc[new_order].reset_index(drop=True)
ribera_debre = grouped_df[grouped_df['COMARCA']== "RIBERA D'EBRE"].iloc[new_order].reset_index(drop=True)
ripolles = grouped_df[grouped_df['COMARCA']== 'RIPOLLES'].iloc[new_order].reset_index(drop=True)
segarra = grouped_df[grouped_df['COMARCA']== 'SEGARRA'].iloc[new_order].reset_index(drop=True)
segria = grouped_df[grouped_df['COMARCA']== 'SEGRIA'].iloc[new_order].reset_index(drop=True)
selva = grouped_df[grouped_df['COMARCA']== 'SELVA'].iloc[new_order].reset_index(drop=True)
solsones = grouped_df[grouped_df['COMARCA']== 'SOLSONES'].iloc[new_order].reset_index(drop=True)
tarragones = grouped_df[grouped_df['COMARCA']== 'TARRAGONES'].iloc[new_order].reset_index(drop=True)
terra_alta = grouped_df[grouped_df['COMARCA']== 'TERRA ALTA'].iloc[new_order].reset_index(drop=True)
urgell = grouped_df[grouped_df['COMARCA']== 'URGELL'].iloc[new_order].reset_index(drop=True)
val_daran = grouped_df[grouped_df['COMARCA']== "VAL D'ARAN"].iloc[new_order].reset_index(drop=True)
valles_occidental = grouped_df[grouped_df['COMARCA']== 'VALLES OCCIDENTAL'].iloc[new_order].reset_index(drop=True)
valles_oriental = grouped_df[grouped_df['COMARCA']== 'VALLES ORIENTAL'].iloc[new_order].reset_index(drop=True)
alt_camp = grouped_df[grouped_df['COMARCA']== 'ALT CAMP'].iloc[new_order].reset_index(drop=True)
alt_emporda = grouped_df[grouped_df['COMARCA']== 'ALT EMPORDA'].iloc[new_order].reset_index(drop=True)
alt_penedes = grouped_df[grouped_df['COMARCA']== 'ALT PENEDES'].iloc[new_order].reset_index(drop=True)
alt_urgell = grouped_df[grouped_df['COMARCA']== 'ALT URGELL'].iloc[new_order].reset_index(drop=True)
alta_ribagorca = grouped_df[grouped_df['COMARCA']== 'ALTA RIBAGORÇA'].iloc[new_order].reset_index(drop=True)
anoia = grouped_df[grouped_df['COMARCA']== 'ANOIA'].iloc[new_order].reset_index(drop=True)
bages = grouped_df[grouped_df['COMARCA']== 'BAGES'].iloc[new_order].reset_index(drop=True)
baix_camp = grouped_df[grouped_df['COMARCA']== 'BAIX CAMP'].iloc[new_order].reset_index(drop=True)
baix_ebre = grouped_df[grouped_df['COMARCA']== 'BAIX EBRE'].iloc[new_order].reset_index(drop=True)
baix_emporda = grouped_df[grouped_df['COMARCA']== 'BAIX EMPORDA'].iloc[new_order].reset_index(drop=True)
baix_llobregat = grouped_df[grouped_df['COMARCA']== 'BAIX LLOBREGAT'].iloc[new_order].reset_index(drop=True)
baix_penedes = grouped_df[grouped_df['COMARCA']== 'BAIX PENEDES'].iloc[new_order].reset_index(drop=True)
barcelones = grouped_df[grouped_df['COMARCA']== 'BARCELONES'].iloc[new_order].reset_index(drop=True)
bergueda = grouped_df[grouped_df['COMARCA']== 'BERGUEDA'].iloc[new_order].reset_index(drop=True)
cerdanya = grouped_df[grouped_df['COMARCA']== 'CERDANYA'].iloc[new_order].reset_index(drop=True)
conca = grouped_df[grouped_df['COMARCA']== 'CONCA DE BARBERA'].iloc[new_order].reset_index(drop=True)
garraf = grouped_df[grouped_df['COMARCA']== 'GARRAF'].iloc[new_order].reset_index(drop=True)
garrigues = grouped_df[grouped_df['COMARCA']== 'GARRIGUES'].iloc[new_order].reset_index(drop=True)
garrotxa = grouped_df[grouped_df['COMARCA']== 'GARROTXA'].iloc[new_order].reset_index(drop=True)
girones = grouped_df[grouped_df['COMARCA']== 'GIRONES'].iloc[new_order].reset_index(drop=True)
maresme = grouped_df[grouped_df['COMARCA']== 'MARESME'].iloc[new_order].reset_index(drop=True)
moianes = grouped_df[grouped_df['COMARCA']== 'MOIANES'].iloc[new_order].reset_index(drop=True)
montsia = grouped_df[grouped_df['COMARCA']== 'MONTSIA'].iloc[new_order].reset_index(drop=True)

montsia
```
