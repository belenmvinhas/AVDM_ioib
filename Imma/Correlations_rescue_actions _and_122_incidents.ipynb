It has been chosen to be studied the relatioship between the rescue actions of the fire departement with the 112 reported incidents with nature-like topics between the years 2018 and 2022. 

## 1. Obtaining data from rescue actions <a class="anchor" id="1"></a>

```python
accions_salvament = pd.read_csv(direc+'Actuacions_en_salvaments_al_medi_natural_dels_Bombers_de_la_Generalitat_20231109.csv')
accions_filtrades= accions_salvament[(accions_salvament["ANY"]<=2022) & (accions_salvament["ANY"] >= 2018)]   #AND & OR |

# show loaded table:
accions_filtrades['TAL_NOM_ALARMA'].unique()
```

```python
accions_filtrades.head()
```

```python
accions_filtrades['NOM_COMARCA'].unique()
```

```python
#Turn "COMARCA" in capital letters

Salvam=accions_filtrades.copy()

from unidecode import unidecode
      
def quitar_acentos(cadena):
    # Reemplazar 'ç' por un marcador temporal
    cadena_con_marcador = cadena.replace('ç', '___cedilla___')
    # Aplicar unidecode y luego revertir el marcador temporal
    cadena_sin_acentos = unidecode(cadena_con_marcador).replace('___cedilla___', 'ç')
    
    return cadena_sin_acentos

Salvam['NOM_COMARCA'] = Salvam['NOM_COMARCA'].apply(quitar_acentos)


Salvam['NOM_COMARCA'] = Salvam['NOM_COMARCA'].str.upper()

Salvam

Salvam['NOM_COMARCA'].unique()
```

```python
#treure totes les columnes que no m'ajuden
#salvam_riu.pop('')
salvam_clean = Salvam.copy()
#Salvam_riu.columns.tolist()

salvam_clean.pop('TAL_COD_ALARMA1') 
salvam_clean.pop('TAL_COD_ALARMA2')
salvam_clean.pop('TGA_NOM_GRUPO')
salvam_clean.pop('CODI_INE')
salvam_clean.pop('NOM_POBLACIO')
salvam_clean.pop('MUNICIPI')
salvam_clean.pop('ACT_NUM_ACTUACIO')
salvam_clean.pop('NOM REGIO')

salvam_clean=salvam_clean[(salvam_clean['VALIDAT'] == 'S') | (salvam_clean['VALIDAT'] == 'SI') ]
salvam_clean['ANY-MES'] = pd.to_datetime(salvam_clean['ANY'].astype(str) + '-' + salvam_clean['MES'].astype(str), format='%Y-%m')
salvam_clean.pop('ANY')
salvam_clean.pop('MES')
salvam_clean.pop('ACT_DAT_ACTUACIO')
salvam_clean.pop('VALIDAT')
salvam_clean.pop('CODI_REGIO')
salvam_clean.pop('ACT_X_UTM')
salvam_clean.pop('ACT_Y_UTM')
salvam_clean.pop('LONGITUD')
salvam_clean.pop('LATITUD')
salvam_clean.pop('GEOREFERÈNCIA')
salvam_clean
```

```python
salvament=salvam_clean.copy()
salvament['CODI_COMARCA'] = salvament['CODI_COMARCA'].astype(str).str.zfill(2)
salvament = salvament.rename(columns={'NOM_COMARCA':"COMARCA"})
salvament
```

### 1.1. Grouped recue actions  <a class="anchor" id="1.1"></a>

```python
#Grouping all recue actions in the same 'COMARCA' in the same month and year
sal = salvament.copy()
sal=pd.DataFrame(sal.groupby(['ANY-MES','CODI_COMARCA','COMARCA'])['ANY-MES'].count())
sal= sal.rename(columns={'ANY-MES':"SALVAMENTS"})
sal = sal.reset_index()
sal_grouped=sal.copy()

# inc_riu_grouped = pd.DataFrame (inc_riu.groupby(['ANY-MES','PROVINCIA','MUNICIPI','COMARCA'])['INCIDENTS'].sum())

sal_grouped
```

## 2. Importing data from 112 incidents <a class="anchor" id="2"></a>

```python
#Importing data and filtering by year
trucades_salvament= pd.read_csv(direc + 'Dades_d_incidents_operatius_gestionats_pel_CAT112_20231110.csv' )
trucades_filtrades= trucades_salvament[(trucades_salvament["ANY"]<2023) & (trucades_salvament["ANY"]>2017)]

trucades_filtrades['COMARCA'].unique()
```

```python
#Filtering by topic
incidents_descartats=['Seguretat','Trànsit','Civisme', 'Altres incidències', 'Fuita','Medi ambient']

incidents = trucades_filtrades[~trucades_filtrades['TIPUS'].isin(incidents_descartats)]
incidents = incidents.reset_index(drop=True)
incidents
```

```python
#Cleaning the data frame

incidents_natura = incidents.copy()

incidents_natura['ANY-MES'] = pd.to_datetime(incidents_natura['ANY'].astype(str) + '-' + incidents_natura['MES'].astype(str), format='%Y-%m')
incidents_natura.pop('ANY')
incidents_natura.pop('MES')
incidents_natura.pop('MUNICIPI')
incidents_natura
```

### 2.1. Grouping incidents at each town each month <a class="anchor" id="2.1"></a>

```python
#Grouping all incidenst in the same town in the same month and year

incidents_natura_grouped = pd.DataFrame (incidents_natura.groupby(['ANY-MES','COMARCA'])['INCIDENTS'].sum())

incidents_natura_grouped= incidents_natura_grouped.reset_index()

incidents_natura_grouped.drop_duplicates()

# result= inc_riu_grouped[inc_riu_grouped['COMARCA']=='BERGUEDA']
# result

incidents_natura_grouped
```

## 3. Merging data <a class="anchor" id="3"></a>

```python
#Merging rescue actions with incidents

merge_sea = pd.merge(incidents_natura_grouped,sal_grouped, on=['COMARCA','ANY-MES'])
merge_sea
```

```python
direct = 'csv_files/'
merge_sea.to_csv(direct + 'Correlations_fire-112.csv', index=False)
```
